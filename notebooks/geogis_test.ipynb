```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cnxn_str = ("Driver={SQL Server};"

            "Server=geosql08;"

            "Database=GeoGIS2010;"

            "Trusted_Connection=yes;")

cnxn = pyodbc.connect(cnxn_str)
```

```python
testquery = ("SELECT TOP (1000) * "
             "FROM [GeoGIS2010].[dbo].[gsf] gsf "
             )
```

```python
data = pd.read_sql(testquery,cnxn)
data
```

```python
anon_key = pd.factorize(df_test["index_key"])[0]
df_test["anon_key"] = anon_key
df_test.drop("index_key", axis=1, inplace=True)
```

```python
df_test
```

```python
df_test = df_test[['anon_key', 'DIAM', 'WPERC']]

df_test.reset_index(inplace=True, drop=True)

fig, ax = plt.subplots()

for key, value in df_test.groupby('anon_key'):
    value.plot(x='DIAM', y='WPERC', ax=ax, label=key, color="#d8d8d8")
    # value.plot(x='DIAM', y='WPERC', ax=ax, label=key) # with colours

plt.xscale('log')
plt.legend('', frameon=False)
```
